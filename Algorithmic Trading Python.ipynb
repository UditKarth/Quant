```python
mro = close.MRO.copy().to_frame() #places the information into a data frame
```

```python
mro
```

```python
mro["lag1"] = mro.shift(periods=1)  #shift one moves the day forwards by 1
```

```python
mro #you can subtract lag1 from mro's usual date to find percentage change
```

```python
mro["dif"] = mro.MRO.sub(mro.lag1)
```

```python
mro
```

```python
mro["% change"] = mro.MRO.div(mro.lag1) #you could shorten the code by computing the percent change at once
```

```python
mro
```

```python
mro["Easy dif"] = mro.MRO.diff(periods=1)
mro
```

```python
mro["Easy %Change"] = mro.MRO.pct_change(periods=1).mul(100)
mro
```

```python
del mro["% change"]
del mro["dif"]
mro
```

```python
mro.rename(columns = {'Easy %Change': 'Daily Change'}, inplace = True)
mro
```

```python
#You can use resample to get specific days
meta = close.META.copy().to_frame()
meta.META.resample("M").last() #gets only the last day of the month
```

```python
ret = meta.pct_change().dropna() #drop NA removes all nan data, making it cleaner
ret = ret.mul(100)
```

```python
ret.plot(kind = "hist", figsize = (15,8), fontsize = 12, bins = 100) #kind = hist results in a histogram being produced
```

```python
ret.mean()
```

```python
daily_var = ret.var() 
daily_var
```

```python
daily_std = np.sqrt(daily_var)
daily_std
```

```python
ret.std()
```

```python
annual = daily_var * 252
annual_std = np.sqrt(annual)
annual_std
```

```python
returns = close.copy()
returns.describe().T #makes a table
```

```python
summary = returns.describe().T.loc[:, ["mean", "std"]]
summary
```

```python
summary.plot.scatter(x = "std", y="mean", figsize = (12,8), s=50, fontsize=15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "std"]+0.002, summary.loc[i, "mean"]+0.002), size = 15) #plots the values(need pyplot)
plt.xlabel("Annual Risk(std)", fontsize = 15)
plt.ylabel("Annual Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 25)
```

## Correlation and Covariance

```python
returns.cov() #shows covariance, or how variables change in the same time frame
```

```python
returns.corr() #shows correlation, or how much two (or more) securities move in relation to one another
```

```python
#The difference between covariance and correlation is that covariance gives you a metric to check how variables change
# in the same way (together) while correlation shows how a change in one variable affects another
```

```python
import seaborn as sns
plt.figure(figsize = (12,8))
sns.set(font_scale=1.4)
sns.heatmap(returns.corr(), cmap = "Reds", annot = True, annot_kws = {"size":15}, vmax = 0.6)
plt.show()
```

The above heatmap shows the correlation between the various companies being examined. You usually want to invest in companies that don't have a high correlation, so that an issue for the sector doesn't result in your entire portfolio tanking.
