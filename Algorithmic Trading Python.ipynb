```python
import yfinance as yf
import pandas as pd
import matplotlib as plt
import numpy as np
```

The library yfinance connects to Yahoo Finance, allowing you to get information from the stock market.

```python
Apple = yf.download("AAPL", start = "2010-01-01", end = "2022-12-31")
```

```python
Apple #yf.download simply saves the information, but by calling the variable we can access the information
```

```python
#If we want to access information for multiple companies at the same time, we can create an array and use the array
# as the input for yf.download
ticker = ["ROBO", "META", "OXY", "MRO"] 
```

```python
stocks = yf.download(ticker, start = "2010-01-01", end = "2022-12-31")
```

```python
stocks
```

```python
stocks.tail() #gets the bottom 5 rows
```

You can make many changes to how pandas incorporates the data:

```python
stocks = pd.read_csv("Stock file", header = [0,1]) # makes more space between heaader and following columns
stocks
```

```python
stocks = pd.read_csv("Stock file", header = [0,1], index_col = [0]) # removes every column excluding the first (0-index)
stocks
```
