## Correlation and Covariance

```python
returns.cov() #shows covariance, or how variables change in the same time frame
```

```python
returns.corr() #shows correlation, or how much two (or more) securities move in relation to one another
```

```python
#The difference between covariance and correlation is that covariance gives you a metric to check how variables change
# in the same way (together) while correlation shows how a change in one variable affects another
```

```python
import seaborn as sns
plt.figure(figsize = (12,8))
sns.set(font_scale=1.4)
sns.heatmap(returns.corr(), cmap = "Reds", annot = True, annot_kws = {"size":15}, vmax = 0.6)
plt.show()
```

The above heatmap shows the correlation between the various companies being examined. You usually want to invest in companies that don't have a high correlation, so that an issue for the sector doesn't result in your entire portfolio tanking.

# Simple Returns and Log Returns

```python
df = pd.DataFrame(index = [2016,2017,2018], data = [100, 50, 99], columns = ["Price"])
df
```

```python
ret1 = df.pct_change().dropna()
ret1.mean()
```

As we can see from the above value, the mean returns can often be misleading; if we take the mean of the returns at face value, we would expect a total return over two years of (100*1.24*1.24 =) 153.76, which is not what we get. Instead, we must find the log returns.

```python
logr = np.log(df/df.shift(1)).dropna()
logr
```

```python
logr.mean()
```

```python
100*np.exp(logr.mean()*2)
#you would do np.exp(logr.mean() * t)
```

# Cumulative Returns and More

```python
apple = yf.download("AAPL", interval = "1wk")
```

```python
apple = apple.Close.to_frame()
```

```python
apple["d_returns"] = np.log(apple.div(apple.shift(1)))
```

```python
apple.dropna(inplace = True)
```

```python
apple
```

```python
apple.d_returns.sum()
```

```python
np.exp(apple.d_returns.sum())
```

```python
apple["cumreturns"] = apple.d_returns.cumsum().apply(np.exp)
```
